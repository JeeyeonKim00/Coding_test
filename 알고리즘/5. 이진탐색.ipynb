# 이진 탐색 알고리즘
- 순차탐색: 리스트 안에 있는 특정한 데이터를 찾기 위해 앞에서부터 데이터를 하나씩 확인하는 방법 (가장 기본. 선택정렬같은..)
- 이진탐색:정렬되어 있는 리스트에서 `탐색 범위를 절반씩 좁혀가며` 데이터를 탐색하는 방법
  - 시작점, 끝점, 중간점을 이용하여 탐색 범위를 설정합니다.

```python
# 재귀 사용
def binary_search(array, target, start, end):
    if start>end:
        return None

    mid = int(start+end)//2
    if array[mid] == target:
        return mid
    elif array[mid] > target:
        return binary_search(array, target, start, mid-1)
    else:
        return binary_search(array, target, mid+1, end)
    
n, target = list(map(int, input().split()))
array = list(map(int, input().split()))

result = binary_search(array, target, 0, n-1)
if result == None:
    print('원소가 존재하지 않습니다.')
else:
    print(result+1) # mid값의 인덱스 반환
```

# 파라메트릭 서치(Parametric Search)
- 최적화 문제를 결정 문제('예'or'아니오') 로 바꾸어 해결하는 기법
  - 최적화 문제? 어떤 함수의 값을 최대한 낮 or 높이는 문제
- 예: 특정한 조건을 만족하는 가장 알맞은 값을 빠르게 찾는 최적화 문제
- 일반적으로 코딩 테스트에서 파라메트릭 서치 문제는 이진 탐색을 이용하여 해결할 수 있다.

```python
# 내 답
N,M = map(int,input().split())
dduk_lst = list(map(int, input().split()))

cut_lst = []

# if min(dduk_lst)
for i in range(min(dduk_lst), max(dduk_lst)):
    lst = [j-i for j in dduk_lst if j-i > 0]
    if sum(lst)<=M:
        cut_lst.append(i)
    # print(f'절단기높이: {i}, 최종 떡 길이: {sum(lst)}')
# print(cut_lst)
print(min(cut_lst))
```

내 생각엔.. start와 end가 만나는게 이진 탐색의 목적인가보오..

```python
# 정답
n,m = list(map(int, input().split(' ')))
array = list(map(int, input().split()))

# 이진 탐색을 위해 처음, 끝점 설정
start = 0
end = max(array)
result = 0

# 이진 탐색 수행
while (start<= end):
    total = 0
    mid = (start+end)//2
    for x in array:
        if x> mid:
            total += x-mid
    # 떡의 양이 부족한 경우, 더 많이 자르기(왼쪽 부분 탐색)
    if total < m:
        end = mid -1
    # 떡의 양이 충분한 경우, 덜 자르기 (오른쪽 부분 탐색)
    else:
        result = mid # 최대한 덜 잘랐을 때가 정답이다. 따라서 result에 기록
        start = mid +1
        
print(result)
```
